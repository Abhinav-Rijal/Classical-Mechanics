# Projectile Motion 
## Horizontal Range
### - $R = \frac{u^2 sin2\theta}{g}$
## Time of Flight 
### - $T = \frac{2u sin\theta}{g}$
## Maximum Height
### - $H = \frac{u^2 sin^2\theta}{2g}$

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Projectile():
    def __init__(self,u,theta,g):
        self.u=u
        self.theta=theta
        self.g=g
    def HorizontalRange(self):
        return (self.u**2)*math.sin(2*self.theta*math.pi/180)/self.g
    def TimeofFlight(self):
        return (self.u*2)*math.sin(self.theta*math.pi/180)/self.g
    def MaximumHeight(self):
        return (self.u*math.sin(self.theta*math.pi/180))**2/(2*self.g)
    def update_theta(self,theta):
        self.theta=theta
        
```

```python
P=Projectile(u=50,theta=0,g=9.8)#initial velocity = 50m/s, g=9.8 m/s^2
R=[]
T=[]
M=[]
A=[]
x=np.arange(0,90+0.1,0.1)
for i in x:
    A.append(i)
    P.update_theta(i)
    r=P.HorizontalRange()
    t=P.TimeofFlight()
    m=P.MaximumHeight()
    R.append(r)
    T.append(t)
    M.append(m)
    
plt.figure(figsize=[11,11])

plt.subplot(2,2,1)
plt.plot(A,R)
plt.xlabel('Angle of Projection (degrees)')
plt.ylabel('Horizontal Range (metres)')
plt.title('Horizontal Range vs Angle of Projection')

plt.subplot(2,2,2)
plt.plot(A,T,color='r')
plt.xlabel('Angle of Projection (degrees)')
plt.ylabel('Time of Flight (seconds)')
plt.title('Time of Flight vs Angle of Projection')

plt.subplot(2,2,3)
plt.plot(A,M,color='g')
plt.xlabel('Angle of Projection (degrees)')
plt.ylabel('Maximum Height (metres)')
plt.title('Maximum Height vs Angle of Projection')
   
```

```python
data={} #empty dictionary
data.update({"Angle":A,"HorizontalRange":R,"TimeofFlight":T,"MaximumHeight":M})
```

```python
u=50
g=9.8
theta=30
X=[]
Y=[]
T=2*u*math.sin(theta*math.pi/180)/g

time=np.arange(0,T,0.1)
for t in time:
    y=u*math.sin(theta*math.pi/180)*t-(g*t**2)/2
    x=u*math.cos(theta*math.pi/180)*t
    Y.append(y)
    X.append(x)
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Projectile Trajectory for u=50 m/s, g=9.8 m/s^2, theta=30 degrees')
```
